## NETFLIX

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")
```

### Drop Null Values

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

### Convert Datetime Format

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['day_added'] = df['date_added'].dt.day
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year
df['year_added'] = df['year_added'].astype('Int64')   # safer int type (handles NaN too)
df['day_added'] = df['day_added'].astype('Int64')
df['month_added'] = df['month_added'].astype('Int64')
```

```python
df.head()
```

# Data Visulaization

#### Movie and TV Shows

```python
mov_tv = netflix['type'].value_counts(normalize=True)*100
mov_tv
```

```python
labels = ['Movie','TV show']
plt.pie(x=mov_tv,labels = labels,autopct='%1.1f%%',startangle=90)

plt.title("Distribution of Movies vs TV Shows", fontsize=14, fontweight='bold')
plt.show()
```

```python
sns.countplot(x='type', data=netflix, palette='Set2')  # use color palette
fig = plt.gcf()
fig.set_size_inches(8, 5)
plt.title('Type')
plt.show()
```

### Rating of shows and movies

```python
count = netflix.rating.value_counts(normalize=True)*100	
```

```python
custom_colors = ["#4CAF50", "#2196F3", "#FFC107", "#F44336", "#9C27B0"]
sns.barplot(data=count,palette=custom_colors)
plt.title("Rating of shows and movies")
plt.xticks(rotation =45)
plt.show()
```

### Relation between Type and Rating

```python
rating_type = netflix.groupby(['rating','type']).size()
rating_type
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=netflix,x='rating',hue='type',palette='Set2')
plt.title("Relation between Type and Rating", fontsize=14, fontweight='bold')
plt.xlabel("Rating")
plt.xticks(rotation=45)
plt.show()
```

### Pie-chart for Rating

```python
rat_pie = netflix['rating'].value_counts(normalize=True)*100
```

```python
rat_pie.index
```

```python
plt.pie(x=rat_pie,autopct='%1.1f%%',labels=rat_pie.index,textprops={'fontsize':6})
plt.title('Pie-chart for Rating')
plt.show()
```
