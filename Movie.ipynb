#### x. Print genre title, maximum movie duration and the count the number of
movies in each genre. (HINT: By using inner join)

```python
movie_genres =pd.read_csv("movie_genres.csv")
movie_genres
```

```python
genres =  pd.read_csv("genres.csv")
genres
```

```python
# Step 1: Merge all three tables using INNER JOIN
merged_df = pd.merge(movies,movie_genres,how ="inner",on = "mov_id")

merged_df= pd.merge(merged_df,genres,how = "inner",on ="gen_id")

result  = merged_df.groupby('gen_title')['mov_time'].agg(max = "max",count ="count")

result.reset_index()
```

### xi. Create a view which should contain the first name, last name, title of the

```python
mv_cs = pd.merge(movies,cast,how= "inner",on ="mov_id")

mv_cs = pd.merge(mv_cs,actor,how="inner", on = "act_id")
view = mv_cs[["act_fname","act_lname","mov_title",'role']]
view.columns = ['first_name', 'last_name', 'movie_title', 'role']
view
```
